```python
import re
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Download stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

# Buat stemmer bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def preprocessing(teks):
    # 1. Lowercase
    teks = teks.lower()
    # 2. Hapus angka & karakter spesial
    teks = re.sub(r'[^a-zA-Z\s]', '', teks)
    # 3. Hapus spasi berlebih
    teks = teks.strip()
    # 4. Hapus stopwords
    kata = teks.split()
    kata = [k for k in kata if k not in stop_words]
    # 5. Stemming (kata dasar)
    kata = [stemmer.stem(k) for k in kata]
    return ' '.join(kata)

# Terapkan ke kolom title dan description
df['title_bersih'] = df['title'].apply(preprocessing)
df['desc_bersih'] = df['description_bersih'].apply(preprocessing)

print("Selesai preprocessing!")
df[['title', 'title_bersih']].head()
```
