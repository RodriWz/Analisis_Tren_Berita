```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/news.db')

# Load tabel news ke DataFrame
df = pd.read_sql_query("SELECT * FROM news", conn)
conn.close()

# Eksplorasi awal
print("Jumlah data:", df.shape)
print("\nKolom:", df.columns.tolist())
print("\n5 data pertama:")
df.head()
```

```python
# Cek data kosong
print(df.isnull().sum())

# Cek contoh isi description (masih HTML)
print(df['description'][0])
```

```python
from bs4 import BeautifulSoup

def bersihkan_html(teks):
    if pd.isna(teks):
        return ""
    soup = BeautifulSoup(teks, "html.parser")
    return soup.get_text(separator=" ").strip()

df['description_bersih'] = df['description'].apply(bersihkan_html)

# Cek hasilnya
print("SESUDAH:", df['description_bersih'][0])
```

```python
df = df.drop(columns=['guid', 'link',])

# Cek kolom yang tersisa
print(df.columns.tolist())
```
